```python
#converting the date into a datetime format
fights['date']=pd.to_datetime(fights['date'],dayfirst=True)
```

```python
#Removing all empty entry rows
fights_cleaned = fights.dropna(how='all' )
```

```python

#encoding fighters
fights["red_fighter"]= fights["R_fighter"].astype("category").cat.codes
fights["blue_fighter"]= fights["B_fighter"].astype("category").cat.codes
```

```python
unique_categories = fights['weight_class'].unique()
unique_categories
```

```python


# removing womens data

rows_to_remove = ['WomenFeatherweight','WomenBantamweight','WomenStrawweight','WomenFlyweight' ]
# Create a boolean mask for rows with women's divisions
mask = ~fights ['weight_class'].isin(rows_to_remove)

# Use boolean indexing to get rows that do not have women's divisions
fights_cleaned = fights_cleaned[mask]
fights_cleaned.shape

```

```python
#remove categorical redundant labels
columns_to_remove = [
 'Referee',
 'location',
 'title_bout',
 'weight_class',
 'B_Stance',
 'R_Stance',
 'B_win_by_Decision_Majority' ,         
 'B_win_by_Decision_Split',                  
 'B_win_by_KO/TKO',                          
 'B_win_by_Submission',                    
 'B_win_by_TKO_Doctor_Stoppage',
 'R_draw',                               
 'R_win_by_Decision_Majority',              
 'R_win_by_Decision_Split'  ,                
'R_win_by_Decision_Unanimous' ,            
'R_win_by_KO/TKO' ,                         
'R_win_by_Submission',                     
'R_win_by_TKO_Doctor_Stoppage',
 ]

# Creating a new dataset without the specified columns
new_dataset =fights_cleaned.drop(columns=columns_to_remove)
new_dataset.head()
```

```python
#Remove fights that ended as draws
rows_to_remove = [ 'Draw']
# Create a boolean mask for rows with Draw
mask = ~new_clean_dataset ['Winner'].isin(rows_to_remove)

# Use boolean indexing to get rows that do not have women's divisions
final_dataset = new_clean_dataset[mask]
final_dataset.shape
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance
label_encoder = LabelEncoder()

# Fit and transform the 'result' column using .loc to avoid SettingWithCopyWarning
final_dataset.loc[:, 'result_encoded'] = label_encoder.fit_transform(final_dataset['Winner'])
```

```python



predictors = final_dataset.columns[4:-1].tolist()



```

```python
train = final_dataset[final_dataset["date"] <"2018-01-01"]
train.head()

```

```python
test = final_dataset[final_dataset["date"] >"2018-06-01"]
test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators = 1000, min_samples_split = 20, random_state=4)
rf.fit(train[predictors], train['result_encoded'])
```

```python
preds= rf.predict(test[predictors])
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
acc= accuracy_score(test['result_encoded'],preds)
```

```python
acc
```

```python
# Import necessary libraries


from sklearn.linear_model import LogisticRegression

# Split the data into training and testing sets

# Create a logistic regression model
model = LogisticRegression()

# Fit the model to the training data
model.fit(train[predictors], train['result_encoded'])

# Make predictions on the test data
y_pred_log = model.predict(test[predictors])

# Evaluate the model
accuracy = metrics.accuracy_score(test['result_encoded'] ,y_pred_log)
precision = metrics.precision_score(test['result_encoded'], y_pred_log)
recall = metrics.recall_score(test['result_encoded'], y_pred_log)
f1_score = metrics.f1_score(test['result_encoded'], y_pred_log)

# Print evaluation metrics
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1_score}')





```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
import numpy as np

# Assuming you have a training dataset named 'train' and predictors are stored in 'predictors'
# Also, assuming 'result_encoded' is the target variable

param_grid = {
    'penalty': ['l1', 'l2'],
    'C': np.logspace(-4, 4, 9),
    'solver': ['liblinear', 'saga']
}

# Create a logistic regression model with an increased max_iter
model = LogisticRegression(max_iter=300)  # You can adjust the value as needed

# Create a grid search object
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=5)

# Fit the grid search to the training data
grid_search.fit(train[predictors], train["result_encoded"])

# Print the best hyperparameters
print("Best Hyperparameters:", grid_search.best_params_)

# Evaluate the model with the best hyperparameters on the test set
best_model = grid_search.best_estimator_
test_accuracy = best_model.score(test[predictors], test["result_encoded"])
print("Test Accuracy with Best Hyperparameters:", test_accuracy)

```

```python

import tensorflow as tf


# Build a simple neural network model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')  # Binary classification, so using 'sigmoid' activation
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train[predictors], train['result_encoded'], epochs=50, batch_size=32, validation_data=(test[predictors],test["result_encoded"] ))

# Evaluate the model on the test set
loss, accuracy = model.evaluate(test[predictors], test["result_encoded"])
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')

```
